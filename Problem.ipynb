### 2.3 เปรียบเทียบ Status ของ Top 5 Locations

```python
top5_locations = complaint_df["location_text"].value_counts().head(5).index
print(top5_locations)
```

```python
top5_df = complaint_df[
    complaint_df["location_text"].isin(top5_locations)
]
```

```python
top5_status_table = pd.crosstab(
    top5_df["location_text"],
    top5_df["status"]
)

top5_status_table
```

```python
ax = top5_status_table.plot(kind="bar", figsize=(10,6))

plt.title("Status Distribution in Top 5 Locations")
plt.ylabel("Number of Cases")
plt.xticks(rotation=45)
plt.legend(title="Status")

# ใส่ตัวเลขบนแท่ง
for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

จากกราฟพบว่า

- ทุกพื้นที่มีสถานะ “รอการช่วยเหลือ” เป็นสัดส่วนสูงที่สุด

- พื้นที่ คอหงส์ หาดใหญ่ สงขลา มีจำนวนเคสรอช่วยเหลือสูงที่สุด (12,395 เคส) สูงกว่าพื้นที่อื่นอย่างชัดเจน

- สถานะ ช่วยเหลือแล้ว และ กำลังดำเนินการช่วยเหลือ มีจำนวนน้อยมากเมื่อเทียบกับเคสรอช่วยเหลือ

ดังนั้น ปัญหาหลักในพื้นที่ที่มีเคสสูง คือ จำนวนเคสรอความช่วยเหลือยังคงสะสมจำนวนมาก และ การจัดการยังไม่ทันกับปริมาณเหตุการณ์ที่เกิดขึ้น

## 3. วิเคราะห์ distribution ของ source (news vs complaint)

```python
news_count = len(news_df)
complaint_count = len(complaint_df)

print("Complaint:", complaint_count)
print("News:", news_count)
```

```python
labels = ["News", "Complaint"]
counts = [news_count, complaint_count]

plt.figure(figsize=(6,4))
bars = plt.bar(labels, counts)

plt.title("Distribution of Source (News vs Complaint)")
plt.ylabel("Number of Cases")

# ใส่ตัวเลขบนแท่ง
for i, v in enumerate(counts):
    plt.text(i, v + (max(counts)*0.01), f"{v:,}", ha='center')

plt.show()
```

จากข้อมูลทั้งหมดพบว่า  
- Complaint = 29,162 เคส  
- News = 602 เคส  

ข้อมูลส่วนใหญ่มาจาก Complaint อย่างชัดเจน (มากกว่า News หลายเท่า)  
แสดงว่าชุดข้อมูลนี้สะท้อนเสียงจากประชาชนเป็นหลัก และอาจมีความเอนเอียงของแหล่งข้อมูล (source bias)
